**Follow-up question:** What does it mean?


## Q2. Which countries lack complete data in our dataset?

Also, how many countries have some missing data?

```python
# Filter the dataframe for 'National 1.5C benchmark' type
national_benchmark_df = df_benchmarks[df_benchmarks["Benchmark type"] == "National 1.5C benchmark"]
print(df_benchmarks["Benchmark type"] == "National 1.5C benchmark")
# Explain: when to use set_index() and what is inplace=True about
national_benchmark_df.set_index(['Country'], inplace=True)

# With this, we select only the columns that contain the years (the country will be the index)
years = [str(year) for year in range(2024, 2031)]
```

```python
# Check if there are missing values in any of the years
# Discuss: is there any advantage to using our index like we introduced before (Country)? Answer: yes, the rest
print(national_benchmark_df[years].isnull())
```

```python
# We need to improve the line above such that we end up with a sum of missing values per row.
# Explain: what axis=1 means and why the output looks different (it's a Series now, no longer a DataFrame).
# Emphasise that Series also have an index, which is the same as the index of the DataFrame.
print(national_benchmark_df[years].isnull().sum(axis=1))
```

```python
# Convert the above to a boolean series
missing_value_rows = national_benchmark_df[years].isnull().sum(axis=1) > 0
print(missing_value_rows)

# Filter the dataframe for rows that contain missing values and get list of countries
print(national_benchmark_df[missing_value_rows].index)
```

## Q3. Prove that the set of countries with missing data is the same for all years.

Use pandas to prove without a doubt that the set of countries with missing data is the same for all years.


```python
# How do we confirm that all of these countries share exactly the same issue?
national_benchmark_df.loc[national_benchmark_df[missing_value_rows].index, years]
```

```python
# How do we robustly confirm that?
national_benchmark_df.loc[national_benchmark_df[missing_value_rows].index, years].isnull().all()
```

# 3. Visualisation 

```python
# Filter the data for a specific country
country = 'United Kingdom'

# Filter the data for the 'National 1.5C benchmark' rows
national_benchmarks = df_benchmarks[df_benchmarks['Benchmark type'] == 'National 1.5C benchmark']

# Filter the data for the specific country
country_benchmarks = national_benchmarks[national_benchmarks['Country'] == country]

# Plot the data
plt.plot(country_benchmarks.columns[7:], country_benchmarks.iloc[0, 7:], marker='o')
plt.xlabel('Year')
plt.ylabel('MtCO₂e')
plt.title(f'National 1.5C benchmark for {country}')
plt.show()
```

```python
# Filter the data for a specific country
country = 'United Kingdom'

# Filter the data for the 'National 1.5C benchmark' rows
national_benchmarks = df_benchmarks[df_benchmarks['Benchmark type'] == 'National 1.5C benchmark']

# Filter the data for the specific country
country_benchmarks = national_benchmarks[national_benchmarks['Country'] == country]

# Plot the data
melted = country_benchmarks.melt(
    id_vars=['Id', 'Country', 'Publication date', 'Emissions metric', 
             'Emissions boundary', 'Units', 'Benchmark type'], 
    var_name='Year', 
    value_name='MtCO₂e')

melted
```

```python
from lets_plot import * # Move this to the top of your notebook

uk_plot = (
    ggplot(melted, aes(x='Year', y='MtCO₂e')) +
    geom_line() +
    ggtitle(f'National 1.5C benchmark for {country}')
)
uk_plot
```
